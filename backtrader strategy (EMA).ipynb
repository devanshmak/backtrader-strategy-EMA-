```python
#importing the libraries

%matplotlib notebook
import backtrader.analyzers as btanalyzers
import matplotlib.pyplot as plt
from datetime import datetime
import backtrader as bt
```

```python
#refer to the backtrader documentation to understand the code below better

class EmaCross(bt.Strategy):  #strategy class
    
    

    def __init__(self):   #'__init__' object has the strategy and indicators
        
        c = bt.ind.EMA(self.data.close, period=10)  
    
        x = bt.ind.EMA(self.data.close, period=20)  

        
        self.crossover = bt.ind.CrossOver(c,x)  #singal condition; i.e :when crossover happens
        

    def next(self):    # 'next' object has conditions for entry and exit 
        
        
        if not self.position:  # not in the market
            if self.crossover > 0:  # if fast crosses slow to the upside
                self.buy()  # enter long

        elif self.crossover < 0:  # in the market & cross to the downside
            self.close()  # close long position
            
            #printing 
        print(len(self))   #no. of bars processed
        print(self.buy)       #buy price
        print(self.position)  #position (in/out)


cerebro = bt.Cerebro()  

cerebro.addsizer(bt.sizers.SizerFix, stake=20) #fixed stake of 20
cerebro.broker.set_cash(50000)   #capital = 50,000
cerebro.broker.setcommission(commission=0.00233)  #standard commissions(Zerodha)


data = bt.feeds.YahooFinanceData(dataname='RELIANCE.NS',   # creating a datafeed
                                 fromdate=datetime(2019, 1, 1),
                                 todate=datetime(2021,1, 1),
                                )

data.addfilter(bt.filters.HeikinAshi(data))   #converting data to heiken ashi candle stick easily with backtrader filter
cerebro.adddata(data)  


cerebro.addstrategy(EmaCross)  #adding our strategy

cerebro.run()  #running the strategy


initial = 50000
final = cerebro.broker.getvalue()  #getting the final value
pnl= final - initial 
x= round(pnl, 2)
print('pnl = ', x)   #printing the pnl
```

```python
#plotting the data

cerebro.plot(style = 'candlestick' , iplot= True, barupfill = 'False', bardownfill = 'False', barup = 'grey', bardown= 'black')
plt.ion()
```
